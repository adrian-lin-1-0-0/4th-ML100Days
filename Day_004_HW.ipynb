```python
import matplotlib.pyplot as plt
plt.plot( app_train['AMT_GOODS_PRICE']+app_train['AMT_INCOME_TOTAL'],app_train['TARGET'], 'b.')
plt.show()
```

#### 可以發現有些極端值
把收入太高的去掉

```python
remove_outliter=app_train['AMT_INCOME_TOTAL']<2500000
new_app_train=app_train[remove_outliter]
plt.plot( new_app_train['AMT_GOODS_PRICE']+new_app_train['AMT_INCOME_TOTAL'],new_app_train['TARGET'], 'b.')
plt.show()
```

可以發現收入加抵押品高於一定值的人幾乎不會有延遲還款問題
